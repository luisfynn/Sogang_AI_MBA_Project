```python
# 서버에서 크롤링을 막은 경우, User-Agent 정보를 이용하여 문제 해결 시도 (www.useragentstring.com에서 확인)
# import urllib.request as req
from bs4 import BeautifulSoup
import bs4.element
import datetime
import requests
```

```python
# 내 user agent string
# Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36

def get_soup_obj(url):
  head = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36'} # 본인 PC에서 User-Agent 정보 검색해서 얻은 값 입력
  res = requests.get(url, headers = head)
  soup = BeautifulSoup(res.text, 'html.parser')
  return soup
```

```python
url = "https://premiumoutlets.co.kr/rpage/store/index/02"
```

```python
soup = get_soup_obj(url)
```

```python
print(soup)
```
