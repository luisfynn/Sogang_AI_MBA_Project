```python
from konlpy.tag import Mecab
mecab = Mecab()
```

```python
mecab.morphs('한글 형태소 분석기(미켑)로 테스트를 해보았습니다. 정상 설치 및 동작이 잘 됩니다.')
```

```python
tokenList = []
tokenSentence = []
speech = df.연설내용
speech = speech.reset_index()

for i in range(0, len(speech)):
  tokens = mecab.morphs(speech.연설내용[i])
  tokenList.append(tokens)
  tokenSentence = list(map(lambda x : " ".join(x), tokenList))

len(tokenSentence)
```

```python
import json
import pandas as pd

with open('/content/drive/Othercomputers/LuisLapTop/WorkSpace/SeleniumExer/TextMiningProj/FinalProject/SentiWord_info.json', encoding='utf-8-sig', mode='r') as f:
  SentiWord_info = json.load(f)

sentiword_dic = pd.DataFrame(SentiWord_info)
```

```python
sentimentDF = pd.DataFrame(columns=("President", "review", "sentiment", "Positive", "Negative"))
idx = 0
sentimentSum = []

for speech in tokenSentence:                        # 문장 하나씩 가져옴
  sentiment = 0                                     # 초기 감성값 0으로 설정
  Positive = 0
  Negative = 0
  for i in range(0, len(sentiword_dic)):            # 감성사전의 모든 단어를 하나씩 선택
    if sentiword_dic.word[i] in speech:             # 연설 문장에 감성 단어가 있는지 확인
      sentiment += int(sentiword_dic.polarity[i])   # 감성단어가 있다면 극성값 합계를 구함.

      if int(sentiword_dic.polarity[i]) >= 0 :
        Positive += int(sentiword_dic.polarity[i])
      else:
        Negative += int(sentiword_dic.polarity[i])
  sentimentDF.loc[idx] = [df.대통령, speech, sentiment, Positive, Negative]        # 문장별 극성값을 데이터프레임으로 쌓음 #df.대통령는 더미 변수
  idx += 1                                          # 다름 문장으로 넘어감
```

```python
df = df.reset_index()
sentimentDF.President = df.대통령
sentimentDF
```
